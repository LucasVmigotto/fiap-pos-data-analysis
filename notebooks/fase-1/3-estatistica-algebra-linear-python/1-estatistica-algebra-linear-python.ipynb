### 1.8. Parte VIII

* Para uma matriz poder ser invertida, ela precisa ser quadrada.
* Para matrizes não quadradas, utiliza-se a pseudo matrix inversa
$$
X = (A^{T} \cdot A)^{-1} \cdot A^{T} \cdot Y
$$

```python
x_1, y_1 = -1, 0
x_2, y_2 = 0, 1
x_3, y_3 = 1, 2
x_4, y_4 = 2, 1

x = [x_1, x_2, x_3, x_4]
y = [y_1, y_2, y_3, y_4]

x, y
```

```python
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(10, 5))

for idx, (el_x, el_y) in enumerate(zip(x, y)):
    plt.scatter(el_x, el_y, 
                color='green',
                label=None if idx != 0 else 'Pares ordenados (x, y)')

plt.plot(x, y, label='Gráfico que liga os pontos')

plt.title('Interpolação de reta sobre pontos')

plt.legend()

plt.grid()

plt.box(False)

plt.show()
```

```python
# Se é uma regressão linear: y = m.x + n

A = np.array([[-1, 1], # 1 = -1 . m + 1 . n
              [0, 1],
              [1, 1],
              [2, 1]])
y = np.array([[y_1], [y_2], [y_3], [y_4]])

A, y
```

$$
X = (A^{T} \cdot A)^{-1} \cdot A^{T} \cdot Y
$$

```python
(m, n) = np.dot(np.dot(np.linalg.pinv(np.dot(A.T, A)), A.T), y)

m, n
```

```python
plt.figure(figsize=(10, 5))

for idx, (el_x, el_y) in enumerate(zip(x, y)):
    plt.scatter(el_x, el_y, 
                color='green',
                label=None if idx != 0 else 'Pares ordenados (x, y)')

plt.plot(x, y, label='Gráfico que liga os pontos')

v_x = np.linspace(-1, 2)

plt.plot(v_x, m * v_x + n, 
         label='Reta otimizada - Método dos Mínimos Quadrados (MMQ)')

plt.title('Interpolação de reta sobre pontos')

plt.legend()

plt.grid()

plt.box(False)

plt.show()
```

$$
a \cdot x^3 + b \cdot x^2 + c \cdot x + d
$$

---

$$
A^{T} \cdot A \cdot x_c = A^{T} \cdot y
$$

$$
X = (A^{T} \cdot A)^{-1} \cdot A^{T} \cdot Y
$$

```python
X = np.array([x ** 3,
              x ** 2,
              x,
              np.ones(len(x))]).T

x, X
```

```python
XT_X = X.T @ X

XT_y = X.T @ y

XT_X, XT_y
```

```python
a, b, c, d = (np.linalg.pinv(XT_X) @ XT_y).ravel()

a, b, c, d
```

```python
plt.figure(figsize=(10, 5))

for idx, (el_x, el_y) in enumerate(zip(x, y)):
    plt.scatter(el_x, el_y, 
                color='green',
                label=None if idx != 0 else 'Pares ordenados (x, y)')

plt.plot(x, y, label='Gráfico que liga os pontos')

v_x = np.linspace(-1, 3)

plt.plot(v_x, a * v_x ** 3 + b * v_x ** 2 + c * v_x + d, 
         label='Reta otimizada - Método dos Mínimos Quadrados (MMQ)')

plt.title('Interpolação de reta sobre pontos')

plt.legend()

plt.grid()

plt.box(False)

plt.show()
```
