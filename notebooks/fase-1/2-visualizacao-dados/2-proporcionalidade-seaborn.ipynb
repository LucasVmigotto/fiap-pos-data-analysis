```python
import pandas as pd
```

```python
PATH_CSV_GASTOS = './data/interim/fase1_parte2_aula1_2-gastos.csv'

df_gastos = pd.read_csv(PATH_CSV_GASTOS)

df_gastos.set_index('Unidade da Federação', inplace=True)

df_gastos.head()
```

```python
PATH_CSV_POPULACAO = './data/interim/fase1_parte2_aula1_2-populacao.csv'

df_populacao = pd.read_csv(PATH_CSV_POPULACAO)

df_populacao.set_index('uf', inplace=True)

df_populacao.head()
```

## Parte I

```python
df_populacao.join(df_gastos)
```

```python
df_populacao_gastos = df_populacao.join(df_gastos)

df_populacao_gastos
```

```python
df_populacao_gastos.plot()
```

```python
df_populacao_gastos.loc[['Ceará', 'Pará']]
```

```python
df_populacao_gastos['gastos_habitante'] = df_populacao_gastos['2021/Ago'] / df_populacao_gastos['populacao']

df_populacao_gastos
```

```python
df_populacao_gastos.plot(kind='bar')
```

```python
df_populacao_gastos['gastos_habitante'].plot(kind='bar')
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(df_populacao_gastos,
                x='populacao',
                y='2021/Ago')
```

```python
plt.figure(figsize=(8, 8))
sns.scatterplot(df_populacao_gastos,
                x='populacao',
                y='gastos_habitante')
```

### Desafios

#### 1 - Para o primeiro gráfico de barras, melhore o layout, ordene por gastos e números de habitantes e adicione marcação para a média

```python
plt.figure(figsize=(20, 6))

sns.barplot(df_populacao_gastos.sort_values(by='gastos_habitante', 
                                            ascending=False),
            x='uf',
            y='gastos_habitante')

plt.hlines(df_populacao_gastos['gastos_habitante'].mean(),
           xmin=0,
           xmax=len(df_populacao_gastos),
           linestyles='--',
           colors='r')

plt.hlines(df_populacao_gastos['gastos_habitante'].median(),
           xmin=0,
           xmax=len(df_populacao_gastos),
           linestyles='dotted',
           colors='y')

plt.annotate('Rio Grande do Norte e Piaui apresentam\ngastos equivalentes à média',
             xy=(11, 5),
             color='r')

plt.annotate('Roraima apresenta gasto equivalente à mediana',
             xy=(13.5, 4.2),
             color='orange')

plt.xticks(rotation=90)

plt.title('Investimentos na Saúde por Habitante por cada Unidade Federativa')
plt.xlabel('Unidade da Federação')
plt.ylabel('Investimento na Saúde por Habitante')

plt.show()
```

## Parte II

```python
PATH_CSV_ORDENADOS = './data/interim/fase1_parte1_aula5-ordenados_total.csv'

df_ordenados = pd.read_csv(PATH_CSV_ORDENADOS,
            index_col=0)

df_ordenados.index = df_ordenados.index.str[3:]

df_ordenados.head()
```

```python
def insere_gastos(gastos, gastos_populacao, mes):
    gastos_populacao[f'gastos_habitante_{mes}'] = gastos[mes] / gastos_populacao['populacao']
    return gastos_populacao
```

```python
# Adiciona Jul
insere_gastos(df_ordenados, 
              df_populacao_gastos, 
              df_ordenados.columns[-2])
```

```python
# Adiciona Jun
insere_gastos(df_ordenados, 
              df_populacao_gastos, 
              df_ordenados.columns[-3])
```

```python
display(df_populacao_gastos)

sns.scatterplot(df_populacao_gastos,
                x='populacao',
                y='gastos_habitante')
sns.scatterplot(df_populacao_gastos,
                x='populacao',
                y='gastos_habitante_2021/Jul')
sns.scatterplot(df_populacao_gastos,
                x='populacao',
                y='gastos_habitante_2021/Jun')
```

### Desafios

#### 1 - Inserir legenda, e melhorar layout do gráfico

```python
fig, ax = plt.subplots(figsize=(12, 6))

sns.scatterplot(df_populacao_gastos,
                x='populacao',
                y='gastos_habitante',
                ax=ax)
sns.scatterplot(df_populacao_gastos,
                x='populacao',
                y='gastos_habitante_2021/Jul',
                ax=ax)
sns.scatterplot(df_populacao_gastos,
                x='populacao',
                y='gastos_habitante_2021/Jun',
                ax=ax)

plt.legend(labels=['Agosto / 2021', 'Julho / 2021', 'Junho / 2021'])

plt.title('Investimento na Área da Saúde por Habitante')
plt.xlabel('População')
plt.ylabel('Investimento por Habitante')

plt.show()
```
